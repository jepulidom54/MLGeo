```python

#uso de paquete de Numpy para generar vectores
x=np.linspace(-10, 10, 21)
y=np.linspace(-20,20,41)


#uso de paquete de Numpy para generar matricesz
[xx, yy]=np.meshgrid(x,y)

#uso de plt funtions para graficar matricez
plt.figure(1)
plt.imshow(xx)
plt.show

plt.figure(2)
plt.imshow(yy)
plt.show

#print(x,y)
```

```python
#subplot fuction ax fig mis variables

fig,ax=plt.subplots(1,2,figsize=(11,8))
ax[0].imshow(xx.T)
ax[0].set_title('XX')
ax[1].imshow(yy.T)
ax[1].set_title('YY')
```

```python
xx_fill=np.zeros((len(x),len(y)))
yy_fill=np.zeros((len(x),len(y)))
for i in range(len(x)):
    for j in range(len(y)):
        xx_fill[i][j]=x[i]
        yy_fill[i][j]=y[j]

```

```python
#subplot fuction ax fig mis variables

fig,ax=plt.subplots(1,2,figsize=(11,8))
ax[0].imshow(xx_fill)
ax[0].set_title('XX-fill')
ax[1].imshow(yy_fill)
ax[1].set_title('YY-fill')
```

```python
f= np.cos(xx.T)*np.sin(yy.T)

plt.imshow(f)
plt.colorbar()
plt.show()
```

```python
#poner un filtro como un cuadrado en el centro de la matriz. si |x|<5 y si |y|<5 esos se vuelven 0
ft_filt=np.zeros((np.shape(f)))
for i in range(0,len(x)):
    for j in range(0,len(y)):
        if np.abs(x[i]) < 5 and np.abs(y[j]) < 5:  
            ft_filt[i][j] = 0
        else:
            ft_filt[i][j] = f[i][j]
```
